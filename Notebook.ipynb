## Práctica 6. Indexación y selección de datos

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Acceso a columnas y estadísticas básicas:


```python
vuelos= pd.read_csv('Data/Vuelos.csv',encoding='latin-1')
```

Imprime las primeras 5 filas del DataFrame.

```python
vuelos.head(5)
```

Calcula la media y la desviación estándar de la columna num_passengers. (Aplicar método mean() y std())

```python
vuelo_media = vuelos['num_passengers'].mean()
print(f"Media de vuelos: {vuelo_media}")
```

```python
vuelos_desviacion = vuelos['num_passengers'].std()
print(f"Desviación de vuelos: {vuelos_desviacion}")
```

¿Cuál es el número máximo de pasajeros registrados? (Aplicar método max()).

```python
max_vuelos= vuelos['num_passengers'].max()
print(f'Maximo de vuelos:{max_vuelos}')
```

# 4.Agregación y agrupación

Agrupa los datos por ruta y calcula la suma total de pasajeros para cada ruta. (aplicar groupby(), sum())

```python
Ruta=vuelos.groupby('route')['num_passengers'].sum().sort_values(ascending=True)
```

¿Cuál es la ruta más popular en términos de pasajeros?

```python
Ruta.tail(1)
```

# 6.Visualización de datos:

Crea un gráfico de barras que muestre la cantidad de pasajeros por día de la semana (flight_day)

```python
demanda_semanal=vuelos.groupby('flight_day')['num_passengers'].mean()
```

```python
demanda_semanal.plot.bar(x=demanda_semanal.index,y=demanda_semanal,rot=0)
plt.title('Demanda por dia de la semana')
plt.xlabel('Dia de la semana')
plt.ylabel('Numero de pasjeros promedio')
plt.show()
```
